```python
import matplotlib.pyplot as plt
import pandas as pd
data = {
    "Column": [
        "Date", "Location", "MinTemp", "MaxTemp", "Rainfall",
        "Evaporation", "Sunshine", "WindGustDir", "WindGustSpeed",
        "WindDir9am", "WindDir3pm", "WindSpeed9am", "WindSpeed3pm",
        "Humidity9am", "Humidity3pm", "Pressure9am", "Pressure3pm",
        "Cloud9am", "Cloud3pm", "Temp9am", "Temp3pm", "RainToday", "RainTomorrow"
    ],
    "Description": [
        "Date of the observation", "Location of the weather station", "Minimum temperature in the 24 hours to 9am",
        "Maximum temperature in the 24 hours to 9am", "Precipitation in the 24 hours to 9am",
        "Class A pan evaporation in the 24 hours to 9am", "Bright sunshine in the 24 hours to midnight",
        "Direction of the strongest wind gust in the 24 hours to midnight", "Speed of the strongest wind gust in the 24 hours to midnight",
        "Direction of the wind at 9am", "Direction of the wind at 3pm", "Speed of the wind at 9am",
        "Speed of the wind at 3pm", "Relative humidity at 9am", "Relative humidity at 3pm",
        "Atmospheric pressure reduced to mean sea level at 9am", "Atmospheric pressure reduced to mean sea level at 3pm",
        "Fraction of sky obscured by cloud at 9am", "Fraction of sky obscured by cloud at 3pm",
        "Temperature at 9am", "Temperature at 3pm", "Did the current day receive precipitation exceeding 1mm in the 24 hours to 9am",
        "Did the next day receive precipitation exceeding 1mm in the 24 hours to 9am"
    ],
    "Units": [
        "N/A", "N/A", "C", "C", "mm",
        "mm", "hours", "compass points", "km/h",
        "compass points", "compass points", "km/h", "km/h",
        "%", "%", "hPa", "hPa", "eighths", "eighths",
        "C", "C", "0/1", "0/1"
    ]
}

df = pd.DataFrame(data)

fig, ax = plt.subplots(figsize=(14, 12)) 

ax.axis('tight')
ax.axis('off')

table = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center')

table.auto_set_font_size(False)
table.set_fontsize(12)  # Increase font size
table.scale(1.2, 1.2)    # Adjust table scale
for (i, j), cell in table.get_celld().items():
    if i == 0:
        cell.set_text_props(weight='bold', color='white')  # Header text properties
        cell.set_facecolor('#4CAF50')  # Header color
    else:
        cell.set_facecolor('#f9f9f9' if i % 2 == 0 else 'white')  # Alternating row colors
    cell.set_edgecolor('black')  # Cell borders
    cell.set_fontsize(12)  # Set font size

table.auto_set_column_width([0, 1, 2])  # Automatically adjust column widths

plt.savefig('weather_data_table.png', bbox_inches='tight', dpi=300)

plt.show()
```
